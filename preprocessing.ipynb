```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import numpy as np
import cv2
```

```python
PATH = "/data/leila/vertebral-fracture/hackathon_2022/train"
CSV_PATH_TRAIN = "/data/leila/vertebral-fracture/hackathon_2022/train/data.csv"
CSV_PATH_TEST = "/data/leila/vertebral-fracture/hackathon_2022/test/data.csv"
```

```python
train_labels = pd.read_csv(CSV_PATH_TRAIN)
```

```python
train_labels_np = np.asarray(train_labels['presence_of_fracture'])
```

```python
plt.bar(["No Fracture", "Fracture"] , [len(train_labels_np) - np.sum(train_labels_np),np.sum(train_labels_np)])
plt.title("Training distribution")
plt.show()
```

```python
print(f"Number of fracture cases {np.sum(train_labels_np)}")
```

```python
print(f"Number of non-fracture cases {len(train_labels_np) - np.sum(train_labels_np)}")
```

## Visualize a fracture vs non-fracture

```python
train_labels.head()
```

```python
sample_fracture = f"{PATH}/{train_labels['img'][0]}"
sample_non_fracture = f"{PATH}/{train_labels['img'][1]}"
```

```python
sample_fracture_img = cv2.imread(sample_fracture)
sample_non_fracture_img = cv2.imread(sample_non_fracture)
```

```python
sample_non_fracture_img.shape
```

```python
plt.imshow(sample_fracture_img / np.max(sample_fracture_img))
plt.title("Fracture")
plt.show()
```

```python
plt.imshow(sample_non_fracture_img / np.max(sample_non_fracture_img))
plt.title("No Fracture")
plt.show()
```
